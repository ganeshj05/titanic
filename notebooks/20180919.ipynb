```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.random.randint(1,11,100)
y = np.random.randint(20,80,100)
```

```python
x = np.random.normal(10,4,100)
y = np.random.normal(10,10,100)
plt.scatter(x,y,c='c',s=20)
plt.show()
```

```python
x = np.linspace(-2*np.pi,2*np.pi,200)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(4*x)
y4 = np.cos(4*x)

#(2 columns, 2 rows, index of graph)
plt.subplot(2,2,1)
plt.plot(x,y1,c='r')
plt.subplot(2,2,2)
plt.plot(x,y2,c='b')
plt.subplot(2,2,3)
plt.plot(x,y3,c='g')
plt.subplot(2,2,4)
plt.plot(x,y4,'-')
plt.legend('test')
```
